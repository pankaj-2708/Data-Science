```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
tips=sns.load_dataset("tips")
```

```python
tips.head()
```

suppose we want to see male smoker/non smoker and female smoker/non smoker gives how much tips on average

```python
tips.groupby(["smoker","sex"])["total_bill"].mean().unstack()
```

```python
# in pivot table we give index and columns as categorical column and values as numerical
tips.pivot_table(index="smoker",columns="sex",values="total_bill")
```

```python
tips.pivot_table(index="smoker",columns="sex",aggfunc="count")
```

```python
tips.pivot_table(index="smoker",columns="sex",values="total_bill",aggfunc="sum")
```

```python
tips.pivot_table(index=["day","time"],columns=["sex","smoker"],values="total_bill")
```

```python
tips.pivot_table(index=["day","time"],columns=["sex","smoker"],values="total_bill")
```

```python
tips.pivot_table(index=["day","time"],columns=["sex","smoker"])
```

```python
tips.pivot_table(index=["day","time"],columns=["sex","smoker"],aggfunc={"size":"sum","tip":"mean","total_bill":"max"})
```

```python
tips.pivot_table(index=["day","time"],columns=["sex","smoker"],values="total_bill",margins=True) # margins=true add a new col and row at end name all which contains sum
```

```python
tips.pivot_table(index=["day"],columns=["sex"],values="total_bill").plot()
# it will plot index as x axis values as y axis and columns as labels
```
