# stack and unstack

stack() pivots columns into rows, effectively transforming a "wide" DataFrame into a "long" DataFrame with a multi-level index. 

Conversely, unstack() pivots rows into columns, converting a "long" DataFrame back to a "wide" DataFrame.

basically unstack convert rows to columns as per multiindexing

and stack convert columns to rows as per multiindexing

```python
df.unstack()
```

```python
s=df.unstack().unstack()
s
```

```python
df.stack()
```

```python

df.stack().stack()
```

```python
death=pd.read_csv("./time_series_covid19_deaths_global.csv")
death.head()
```

```python
confirm=pd.read_csv("./time_series_covid19_confirmed_global.csv")
confirm.head()
```

```python
df1=death.melt(id_vars=["Country/Region","Province/State","Lat","Long"],var_name="date",value_name="death")
df2=confirm.melt(id_vars=["Country/Region","Province/State","Lat","Long"],var_name="date",value_name="confirmed")
```

```python
df1["Confirmed"]=df2["confirmed"]
```

```python
df1.sort_values("Country/Region",inplace=True)
```

```python
df1.reset_index(inplace=True,drop=True)
```

```python
ml2=pd.MultiIndex.from_product([death["Country/Region"].values,df2["date"].unique().tolist()])
```

```python
df1.set_index(ml2,inplace=True)
```

```python
df1.drop(columns=["Lat","Long","Province/State","Country/Region","date"],inplace=True)
```

```python
df1.to_csv("modified_covid_death_and_confirmed_data.csv")
```
