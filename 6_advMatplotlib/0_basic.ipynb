```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# adding annoations

# we can add size,text,colour to points of scatter plot 

x = [10, 20, 30, 40, 50]
y = [5, 15, 25, 35, 45]
sizes = [50, 200, 300, 400, 100]            # Different sizes for each point
colors = ['red', 'green', 'blue', 'purple', 'orange']  # Different colors

# Scatter plot with color and size
plt.scatter(x, y, s=sizes, c=colors, alpha=0.6)

# Add annotations
for i in range(len(x)):
    plt.text(x[i], y[i], f"({x[i]}, {y[i]})", fontsize=9, ha='right')

# Titles and labels
plt.title("Scatter Plot with Annotations")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.grid(True)

# adding lines to graph

plt.axhline(30,color='red')
plt.axvline(30,color='red')
plt.show()
```

# subplots

```python
# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.exp(-x / 3) * np.sin(2 * x)
y4 = np.random.randn(100).cumsum()

# Create 2x2 subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8), sharex=False)

# Plot 1 - Line plot
axs[0, 0].plot(x, y1, color='blue', linestyle='-', marker='o')
axs[0, 0].set_title("Sine Wave")
axs[0, 0].set_ylabel("Amplitude")
axs[0, 0].grid(True)

# Plot 2 - Cosine
axs[0, 1].plot(x, y2, color='green', linestyle='--', linewidth=2)
axs[0, 1].set_title("Cosine Wave")
axs[0, 1].grid(True)

# Plot 3 - Damped Sine
axs[1, 0].plot(x, y3, color='red')
axs[1, 0].set_title("Damped Sine")
axs[1, 0].set_xlabel("Time")
axs[1, 0].set_ylabel("Amplitude")
axs[1, 0].grid(True)

# Plot 4 - Cumulative Random
axs[1, 1].bar(range(len(y4)), y4, color='orange')
axs[1, 1].set_title("Cumulative Random")
axs[1, 1].set_xlabel("Index")
axs[1, 1].grid(True)

# Add a main title for the whole figure
fig.suptitle("Multiple Subplots with Major Features", fontsize=14)

```

# 3d surface plots

```python
X=np.linspace(-10,10,100)
Y=np.linspace(-10,10,100)
```

```python
XX,YY=np.meshgrid(X,Y)
XX.shape
```

```python
z=XX**2+YY**2
z.shape
```

```python
ax=plt.subplot(projection='3d')

ax.plot_surface(XX,YY,z)
```

# contour plot

```python
plt.contour(XX,YY,z)
```

```python
plt.contourf(XX,YY,z)
```
